```python
print(df.isnull().sum())
sns.heatmap(df.isnull(), cbar=False, cmap='coolwarm')
plt.title("Missing Values Heatmap")
plt.show()
```

```python
duplicates = df.duplicated().sum()
print(f"\nðŸ” Total Duplicate Records: {duplicates}")
```

```python
# Remove duplicates if any
if duplicates > 0:
    df.drop_duplicates(inplace=True)
    print("Duplicates removed. New shape:", df.shape)
```

```python
print(df.describe(include='all'))
```

```python
# 6. Unique Values in Categorical Columns
categorical_cols = df.select_dtypes(include=['object']).columns
print("\nðŸ§¬ Unique Values in Categorical Columns:")
for col in categorical_cols:
    print(f"{col}: {df[col].unique()}")
```

```python
# 7. Correlation Matrix (Numerical)
numeric_cols = df.select_dtypes(include=[np.number])
if not numeric_cols.empty:
    corr = numeric_cols.corr()
    plt.figure(figsize=(10,6))
    sns.heatmap(corr, annot=True, cmap='viridis')
    plt.title("Correlation Heatmap")
    plt.show()
```

```python
# Histogram for all numerical columns
df.hist(bins=20, figsize=(12, 10), color='skyblue', edgecolor='black')
plt.suptitle("Histogram of Numerical Features")
plt.show()
```

```python
# Pairplot (if dataset size allows)
if df.shape[1] <= 10:
    sns.pairplot(df)
    plt.suptitle("Pair Plot", y=1.02)
    plt.show()
    
    
```

```python
# Boxplots
for col in numeric_cols.columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=df[col], color='lightgreen')
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python

# Countplots for categorical variables
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=df[col], palette='Set2')
    plt.title(f"Count Plot of {col}")
    plt.xticks(rotation=45)
    plt.show()
```
