```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("car_data.csv")   
df

```

```python
#Bar Chart 
plt.figure(figsize=(6,4))
sns.barplot(x="Fuel_Type", y="Selling_Price", data=df, estimator=np.mean)
plt.title("Average Selling Price by Fuel Type")
plt.show()

```

```python
# 2. Pie Chart - Distribution of Fuel Type
plt.figure(figsize=(5,5))
df["Fuel_Type"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90, cmap="Set2")
plt.ylabel("")
plt.title("Fuel Type Distribution")
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(x="Present_Price", y="Selling_Price", hue="Fuel_Type", data=df)
plt.title("Present Price vs Selling Price")
plt.show()
```

```python
#Line Plot
plt.figure(figsize=(6,4))
sns.lineplot(x="Year", y="Selling_Price", data=df, estimator=np.mean)
plt.title("Trend of Selling Price by Year")
plt.show()
```

```python
#Pair Plot
sns.pairplot(df[["Year","Selling_Price","Present_Price","Kms_Driven"]], diag_kind="hist")
plt.show()
```

```python

#Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x="Transmission", y="Selling_Price", data=df)
plt.title("Selling Price by Transmission")
plt.show()
```

```python
#Heatmap
selected_columns = ['Year','Selling_Price','Present_Price']

corr = [selected_columns].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr,annot = True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()


```
