```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("shopping_trends.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
# histogram plot of age vs frequency
df['Age'].plot(kind = "hist", edgecolor = "blue")
plt.xlabel("Age")
plt.title("Age vs Frequency")
plt.show()
```

```python
# bar plot of gender vs count
df["Gender"].value_counts().plot(kind = 'bar', color = ['red', 'green'])
plt.title("Gender vs count")
plt.grid(alpha = 0.5)
plt.show()
```

```python
# Average of Purchase amount by category
df.groupby("Category")['Purchase Amount (USD)'].mean().sort_values(ascending = False).tail(5)
```

```python
# Review Rating in Descending order
df.sort_values("Review Rating", ascending = False, inplace = True)

```

```python
df
```

```python
# Count of Subscription Status
df['Subscription Status'].value_counts().plot(kind = 'pie', autopct = '%1.1f%%' )
plt.title("Pie chart of Subscription Status Count")
plt.show()
```

```python
df.Color.unique()
```

```python
plt.figure(figsize= (10, 20))
df.boxplot(column ='Purchase Amount (USD)',by = "Frequency of Purchases")
plt.title("Box Plot")
plt.show()
```

```python
# Average of Purchased Amount based on different size
df.groupby("Size")["Purchase Amount (USD)"].mean()
df
```

```python
df.groupby("Payment Method")["Purchase Amount (USD)"].sum().sort_values(ascending = False).plot(kind = "line")
plt.title("Sum of Purchased Method in descending order based on Payement type ")
plt.show()
```

```python
# Median of Previous Purchases groupedby Frequency of Purchases
df.groupby("Frequency of Purchases")['Previous Purchases'].median()
```

```python
# Count of Season
df['Season'].value_counts().plot(kind = 'bar')
plt.title("Season Count")
plt.grid()
plt.show()
```

```python
df['Category'].unique()
```

```python
df.groupby("Category")["Purchase Amount (USD)"].sum().plot(kind = "line")
plt.title("Line plot for Purchased Amount Groupedby Category")
plt.show()
```

```python
df.groupby("Category")["Purchase Amount (USD)"].sum().reset_index()
```

```python
df.groupby("Category")["Purchase Amount (USD)"].sum().reset_index().plot(
    kind="scatter",
    x="Category",
    y="Purchase Amount (USD)",
    marker = "D"
)
plt.show()
```
