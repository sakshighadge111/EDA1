```python
df.drop_duplicates(inplace=True)
#df = df.drop_duplicates()
```

```python
# removing the column
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('/,...,_,-,#,$,&,!')
df
```

```python
# phone number to standard format ddd-ddd-dddd
df['Phone_Number'] = df['Phone_Number'].replace('[^a-zA-Z0-9]', '', regex = True)
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace("nan--",'')
df['Phone_Number']= df['Phone_Number'].str.replace("Na--",'')
```

```python
df
```

```python
df['Paying Customer']= df['Paying Customer'].replace('Y', 'Yes')
df['Paying Customer']= df['Paying Customer'].replace('N', 'No')
df['Paying Customer'] = df['Paying Customer'].replace('N/a', "")
```

```python
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('N', 'No')
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('Y', 'Yes')
```

```python
df
```

```python
#df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('NaN','')
# That happens because in pandas, missing values are stored as np.nan (a special floating-point "Not a Number"), not as the string "NaN"
df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna('')
df
```

```python
df[['Street_Address','State','Zip_Code']] = df['Address'].str.split(',',expand = True)
```
